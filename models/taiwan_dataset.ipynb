feature engineering


```python
pay_status_cols = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']

# Calculate delay penalty score
delay_penalty = df[pay_status_cols].clip(0, 9).sum(axis=1)  # Sum all delays

# Convert to percentiles (this will spread out the scores)
df['payment_history'] = (1 - delay_penalty.rank(pct=True)).round(2)
```

```python
df['avg_bill']=df[['BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6']].mean(axis=1)
df['credit_utilization_ratio']=(df['avg_bill']/df['LIMIT_BAL']).clip(upper=1)
df['credit_utilization_ratio']=pd.to_numeric(1-df['credit_utilization_ratio']).round(2)
df
```

```python
bill_cols = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']
df['number_of_credit_accounts'] = ((df[bill_cols] > 0).sum(axis=1) / 6).round(2)
# Convert to percentile ranks - this worked great for payment_history
df['number_of_credit_accounts'] = df['number_of_credit_accounts'].rank(pct=True).round(2)
```

```python
edu_map = {1: 0.85, 2: 0.8, 3: 0.75, 4: 0.70, 5: 0.7, 6: 0.7, 0: 0.7}
df['education_level'] = df['EDUCATION'].map(edu_map)
df
```

```python
age_factor = df['AGE'].apply(lambda x: min(1.0, (x - 18) / 30))
activity_factor = ((df[bill_cols] > 0).sum(axis=1) / 6)    
df['credit_history_length'] = (
    (age_factor * 0.7) + (activity_factor * 0.3)
).round(2)
df
```

```python
df['weighted_score'] = (
    df['payment_history'] * 0.35 +
    df['credit_utilization_ratio'] * 0.30 +
    df['credit_history_length'] * 0.15 +      # New factor
    df['number_of_credit_accounts'] * 0.10 +
    df['education_level'] * 0.10
)
df['credit_score'] = df['weighted_score'] * 700 + 300
df
```

```python
def numb_outliers(df):
    for i in df.columns:
        if df[i].dtype=='float64' or df[i].dtype=='int64':
            q1=df[i].quantile(.25)
            q3=df[i].quantile(.75)
            iqr=q3-q1
            ub=q3+1.5*iqr
            lb=q1-1.5*iqr
            print('The number of outliers in',i,'is',df[(df[i]>ub) | (df[i]<lb)][i].count())

numb_outliers(df)
```

```python
df['target'] = df['default_payment_next_month'].map({0: 0, 1: 1})
df['target'] = df['target'].astype(int)

y = df['target']
print(f"New target dtype: {y.dtype}")
print(f"Value counts:\n{y.value_counts()}")

x = df[['payment_history', 
        'credit_utilization_ratio', 
        'credit_history_length', 
        'number_of_credit_accounts', 
        'education_level']]
print(x)
```

```python
from sklearn.model_selection import cross_val_score,train_test_split
x_train,x_temp,y_train,y_temp=train_test_split(x,y,test_size=0.4,stratify=y)
x_val, x_test, y_val, y_test = train_test_split(x_temp, y_temp, test_size=0.5, stratify=y_temp)
x_test
y
```
