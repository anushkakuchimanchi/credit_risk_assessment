```python
def outliers(df):
    for i in df.columns:
        if df[i].dtype=='int64' or df[i].dtype=='float64':
            sns.boxplot(df[i])
            plt.show()
```

```python
outliers(df)
```

```python
def numb_outliers(df):
    for i in df.columns:
        if df[i].dtype=='float64' or df[i].dtype=='int64':
            q1=df[i].quantile(.25)
            q3=df[i].quantile(.75)
            iqr=q3-q1
            ub=q3+1.5*iqr
            lb=q1-1.5*iqr
            print('The number of outliers in',i,'is',df[(df[i]>ub) | (df[i]<lb)][i].count())

numb_outliers(df)
```

```python
education_map = {'High School':1,'Bachelor':2,'Master':3,'PhD':4}
employment_status = {'Unemployed':0,'Employed':1,'Self-Employed': 2}

df['Education Level'] = df['Education Level'].map(education_map)
df['Employment Status'] = df['Employment Status'].map(employment_status)

credit_scores = []
ratings=[]
for index, row in df.iterrows():
    payment_history = row['Payment History']
    credit_utilization_ratio = row['Credit Utilization Ratio']
    number_of_credit_accounts = row['Number of Credit Accounts']
    education_level = row['Education Level']
    employment_status = row['Employment Status']
   
    credit_score = (payment_history*0.35) + (credit_utilization_ratio*0.3) + (number_of_credit_accounts*0.15) + (education_level*0.1) + (employment_status*0.1)
    credit_scores.append(credit_score)
    if credit_score<650:
        ratings.append(0)
    else:
        ratings.append(1)

df['Credit Score'] = credit_scores
df['rating']=ratings
df.head()
```

```python
df.dtypes
```

```python
gender_map = {'Male': 0, 'Female': 1}
df['Gender'] = df['Gender'].map(gender_map)
marital_map = {'Married': 0, 'Single': 1,'Divorced': 2}
df['Marital Status'] = df['Marital Status'].map(marital_map)
loan_map = {'Personal Loan': 0, 'Auto Loan': 1,'Home Loan': 2}
df['Type of Loan'] = df['Type of Loan'].map(loan_map)
df.head()
```

```python
y=np.array(df['rating'])
x=np.array(df.drop('rating',axis=1))
df.head()
```

```python
from sklearn.model_selection import cross_val_score,train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1', 'l2'],'C':[1,2,3,4,5,6,10,20,30,50,100],'max_iter':[100,200,300]}
```

```python
classifier_g=GridSearchCV(classifier,param_grid=parameter,scoring='accuracy')
```

```python
classifier_g.fit(x_train,y_train)
```

```python
classifier_g.best_params_
```

```python
y_pred=classifier_g.predict(x_test)
y_test==y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix,roc_curve,roc_auc_score,precision_recall_curve, accuracy_score,precision_score,recall_score,f1_score
print(classification_report(y_test,classifier_g.predict(x_test)))
```

Advanced Model-XGBOOST regressor

```python

rate = []
for index, row in df.iterrows():
    credit_scor = row['Credit Score']
    if credit_scor < 580:
        rate.append(0)
    elif credit_scor < 670:  
        rate.append(1)
    elif credit_scor < 740: 
        rate.append(2)  
    elif credit_scor < 800: 
        rate.append(3)
    else: 
        rate.append(4)

df['rating'] = rate
df['rating'].value_counts()
```

```python
y_n=np.array(df['rating'])
x_n=np.array(df.drop(['rating','Credit Score'],axis=1))
#x_n.head()
```

```python
x_train_n,x_temp,y_train_n,y_temp=train_test_split(x_n,y_n,test_size=0.4,stratify=y)
x_val, x_test_n, y_val, y_test_n = train_test_split(x_temp, y_temp, test_size=0.5, stratify=y_temp)
x_test_n.shape
```

```python
y_test_n
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train_n)
x_val_scaled = scaler.transform(x_val)
x_test_scaled = scaler.transform(x_test_n)
```

```python
import xgboost as xgb
from xgboost import XGBClassifier

param_grid={'n_estimators':[100, 200, 300],'max_depth': [3, 4, 5, 6],'learning_rate': [0.01, 0.1, 0.2],'subsample': [0.8, 0.9, 1.0],'colsample_bytree': [0.8, 0.9, 1.0]}
xgb_model = XGBClassifier(objective='multi:softprob',eval_metric='logloss',random_state=42,n_jobs=-1)
xgb_grid=GridSearchCV(xgb_model,param_grid,cv=3,scoring='accuracy',n_jobs=-1,verbose=1)
```

```python
xgb_grid.fit(X_train_scaled,y_train_n)
```

```python
xgb_grid.best_params_
```

```python
y_pred_n=xgb_grid.predict(x_test_scaled)
y_test_n==y_pred_n
```

```python
val_pred=xgb_grid.predict(x_val_scaled)
y_val==val_pred
```

```python

print(classification_report(y_test_n,xgb_grid.predict(x_test_scaled)))
```
